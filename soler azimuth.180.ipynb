```python
#### import math
import numpy as np
import dash
from dash import dcc, html, Input, Output, State
import dash_bootstrap_components as dbc
import dash_table
import datetime

# ============= (1) نموذج جدول معاملات (Polynomials) ============
poly_data = [
    {
        "month": 1,
        "RA_a0": 6.611038, "RA_a1": 2.102738, "RA_a2": 0, "RA_a3": 0, "RA_a4": 0,
        "Dec_a0": -23.13125, "Dec_a1": 2.20385, "Dec_a2": 3.99089, "Dec_a3": -0.26584, "Dec_a4": -0.10265,
        "E_a0": 11.956617, "E_a1": -0.254556, "E_a2": 0.027385, "E_a3": 0.065938, "E_a4": -0.018779,
        "SD_a0": 0.27117, "SD_a1": -0.00051, "SD_a2": 0, "SD_a3": 0, "SD_a4": 0,
    },
    {
        "month": 2,
        "RA_a0": 8.648065, "RA_a1": 2.102713, "RA_a2": 0, "RA_a3": 0, "RA_a4": 0,
        "Dec_a0": -17.58293, "Dec_a1": 8.81239, "Dec_a2": 2.72612, "Dec_a3": -0.713, "Dec_a4": 0.03897,
        "E_a0": 11.77912, "E_a1": -0.083925, "E_a2": 0.110637, "E_a3": 0.009052, "E_a4": -0.013067,
        "SD_a0": 0.27066, "SD_a1": -0.0017, "SD_a2": 0, "SD_a3": 0, "SD_a4": 0,
    },
    {
        "month": 3,
        "RA_a0": 10.553648, "RA_a1": 2.1027, "RA_a2": 0, "RA_a3": 0, "RA_a4": 0,
        "Dec_a0": -7.86213, "Dec_a1": 12.10851, "Dec_a2": 1.02962, "Dec_a3": -0.68811, "Dec_a4": 0.05899,
        "E_a0": 11.791837, "E_a1": 0.100397, "E_a2": 0.068161, "E_a3": -0.019244, "E_a4": -0.005243,
        "SD_a0": 0.26912, "SD_a1": -0.00233, "SD_a2": 0, "SD_a3": 0, "SD_a4": 0,
    },
    {
        "month": 4,
        "RA_a0": 12.590639, "RA_a1": 2.10271, "RA_a2": 0, "RA_a3": 0, "RA_a4": 0,
        "Dec_a0": 4.26066, "Dec_a1": 12.37858, "Dec_a2": -0.61676, "Dec_a3": -0.61431, "Dec_a4": 0.05314,
        "E_a0": 11.930946, "E_a1": 0.159559, "E_a2": -0.018641, "E_a3": -0.021105, "E_a4": -0.000304,
        "SD_a0": 0.26683, "SD_a1": -0.00234, "SD_a2": 0, "SD_a3": 0, "SD_a4": 0,
    },
    {
        "month": 5,
        "RA_a0": 14.561929, "RA_a1": 2.102729, "RA_a2": 0, "RA_a3": 0, "RA_a4": 0,
        "Dec_a0": 14.85844, "Dec_a1": 9.77582, "Dec_a2": -2.05906, "Dec_a3": -0.5655, "Dec_a4": 0.08234,
        "E_a0": 12.046516, "E_a1": 0.068294, "E_a2": -0.081569, "E_a3": -0.001079, "E_a4": 0.003795,
        "SD_a0": 0.26458, "SD_a1": -0.00175, "SD_a2": 0, "SD_a3": 0, "SD_a4": 0,
    },
    {
        "month": 6,
        "RA_a0": 16.598948, "RA_a1": 2.102742, "RA_a2": 0, "RA_a3": 0, "RA_a4": 0,
        "Dec_a0": 21.95475, "Dec_a1": 4.49412, "Dec_a2": -3.26719, "Dec_a3": -0.28433, "Dec_a4": 0.11682,
        "E_a0": 12.038481, "E_a1": -0.078539, "E_a2": -0.065142, "E_a3": 0.04103, "E_a4": -0.003825,
        "SD_a0": 0.26285, "SD_a1": -0.00073, "SD_a2": 0, "SD_a3": 0, "SD_a4": 0,
    },
    {
        "month": 7,
        "RA_a0": 18.57027, "RA_a1": 2.102737, "RA_a2": 0, "RA_a3": 0, "RA_a4": 0,
        "Dec_a0": 23.15236, "Dec_a1": -1.99345, "Dec_a2": -3.49721, "Dec_a3": 0.2288, "Dec_a4": 0.05073,
        "E_a0": 11.938448, "E_a1": -0.105102, "E_a2": 0.030157, "E_a3": 0.045052, "E_a4": -0.014413,
        "SD_a0": 0.26212, "SD_a1": 0.00044, "SD_a2": 0, "SD_a3": 0, "SD_a4": 0,
    },
    {
        "month": 8,
        "RA_a0": 20.607298, "RA_a1": 2.102717, "RA_a2": 0, "RA_a3": 0, "RA_a4": 0,
        "Dec_a0": 18.19182, "Dec_a1": -7.94043, "Dec_a2": -2.56329, "Dec_a3": 0.50791, "Dec_a4": -0.02149,
        "E_a0": 11.893203, "E_a1": 0.027675, "E_a2": 0.086688, "E_a3": 0.002457, "E_a4": -0.01004,
        "SD_a0": 0.26254, "SD_a1": 0.00153, "SD_a2": 0, "SD_a3": 0, "SD_a4": 0,
    },
    {
        "month": 9,
        "RA_a0": 22.644303, "RA_a1": 2.1027, "RA_a2": 0, "RA_a3": 0, "RA_a4": 0,
        "Dec_a0": 8.53684, "Dec_a1": -11.55544, "Dec_a2": -1.16098, "Dec_a3": 0.50663, "Dec_a4": -0.0094,
        "E_a0": 11.994767, "E_a1": 0.16605, "E_a2": 0.045959, "E_a3": -0.02545, "E_a4": -0.003454,
        "SD_a0": 0.26403, "SD_a1": 0.00226, "SD_a2": 0, "SD_a3": 0, "SD_a4": 0,
    },
    {
        "month": 10,
        "RA_a0": 0.615582, "RA_a1": 2.102706, "RA_a2": 0, "RA_a3": 0, "RA_a4": 0,
        "Dec_a0": -2.90661, "Dec_a1": -12.42961, "Dec_a2": 0.2725, "Dec_a3": 0.53552, "Dec_a4": 0.01351,
        "E_a0": 12.167194, "E_a1": 0.173351, "E_a2": -0.036286, "E_a3": -0.032521, "E_a4": 0.002127,
        "SD_a0": 0.26618, "SD_a1": 0.00241, "SD_a2": 0, "SD_a3": 0, "SD_a4": 0,
    },
    {
        "month": 11,
        "RA_a0": 2.652576, "RA_a1": 2.102732, "RA_a2": 0, "RA_a3": 0, "RA_a4": 0,
        "Dec_a0": -14.19324, "Dec_a1": -10.34392, "Dec_a2": 1.95147, "Dec_a3": 0.62944, "Dec_a4": -0.03537,
        "E_a0": 12.273398, "E_a1": 0.018906, "E_a2": -0.107746, "E_a3": -0.019965, "E_a4": 0.012345,
        "SD_a0": 0.26855, "SD_a1": 0.00188, "SD_a2": 0, "SD_a3": 0, "SD_a4": 0,
    },
    {
        "month": 12,
        "RA_a0": 4.623888, "RA_a1": 2.10275, "RA_a2": 0, "RA_a3": 0, "RA_a4": 0,
        "Dec_a0": -21.68415, "Dec_a1": -5.14283, "Dec_a2": 3.54166, "Dec_a3": 0.44125, "Dec_a4": -0.15409,
        "E_a0": 12.189505, "E_a1": -0.195283, "E_a2": -0.092461, "E_a3": 0.036044, "E_a4": 0.00482,
        "SD_a0": 0.27037, "SD_a1": 0.00081, "SD_a2": 0, "SD_a3": 0, "SD_a4": 0,
    },
]




       


# ============= (2) بعض الدوال المساعدة ============

def hours_mod24(h):
    """دالة لضبط الساعة لتكون بين 0 و 24"""
    while h < 0:
        h += 24
    while h >= 24:
        h -= 24
    return h

def normalize_pm180(a):
    """دالة لضبط الزوايا بين -180° و 180°"""
    while a > 180:
        a -= 360
    while a < -180:
        a += 360
    return a

def normalize_0_360(a):
    """دالة لضبط الزوايا بين 0° و 360°"""
    return a % 360

def dms_to_decimal(dg, mn, sc):
    """دالة لتحويل القيم من DMS (درجات، دقائق، ثواني) إلى درجات عشرية"""
    if any(x is None for x in [dg, mn, sc]):
        return None
    try:
        return float(dg) + float(mn) / 60 + float(sc) / 3600
    except:
        return None

def decimal_to_dms(dd):
    """دالة لتحويل القيم من درجات عشرية إلى DMS"""
    if dd is None:
        return "N/A"
    sign = -1 if dd < 0 else 1
    dd = abs(dd)
    deg = int(dd)
    rem = (dd - deg) * 60
    mins = int(rem)
    secs = (rem - mins) * 60
    deg = deg * sign
    return f"{deg:03d}° {mins:02d}' {secs:05.2f}\""

def hms_to_hours(h, m, s):
    """تحويل ساعات ودقائق وثواني إلى ساعات عشرية"""
    try:
        h = float(h)
        m = float(m)
        s = float(s)
        return h + m / 60.0 + s / 3600.0
    except (ValueError, TypeError):
        return None  # إرجاع None 

def seconds_to_hms(seconds):
    """تحويل الثواني إلى ساعات ودقائق وثواني"""
    try:
        seconds = float(seconds)
        hours = int(seconds // 3600)
        minutes = int((seconds % 3600) // 60)
        remaining_seconds = int(seconds % 60)
        return hours, minutes, remaining_seconds
    except (ValueError, TypeError):
        return None

def time_to_seconds(h, m, s):
    """تحويل ساعات ودقائق وثواني إلى ثواني"""
    try:
        h = float(h)
        m = float(m)
        s = float(s)
        if h < 0 or m < 0 or s < 0:
            raise ValueError("Hours, minutes, and seconds must be non-negative.")
        return h * 3600 + m * 60 + s
    except:
        return None

def poly_value(a0, a1, a2, a3, a4, x):
    """دالة لحساب القيم البولينية باستخدام المعاملات"""
    return a0 + a1*x + a2*x**2 + a3*x**3 + a4*x**4


# ============= (3) دالة لحساب Azimuth Sun بشكل صحيح ============
def get_azimuth(lat_dd, lon_dd, month, day_of_month, local_hh, utc_corr_hh, poly_data):
    """
    حساب أزيموث الشمس وارتفاعها (Altitude) بناءً على:
    - الموقع (lat_dd, lon_dd)
    - اليوم (month, day_of_month)
    - التوقيت المحلي (local_hh)
    - التصحيح الزمني UTC (utc_corr_hh)
    - معاملات البولينومي (poly_data)
    """
    # 1) حساب وقت GMT/UTC
    #gmt_hh = local_hh - utc_corr_hh
    gmt_hh = local_hh - utc_corr_hh

    if gmt_hh < 0:
        gmt_hh += 24
        day_of_month -= 1
    elif gmt_hh >= 24:
        gmt_hh -= 24
        day_of_month += 1
    else:
        # 0 <= gmt_hh < 24
        # لا نقوم بتعديل الوقت أو اليوم
        pass

    # 2) نحصل على صف معاملات الشهر
    row = next((r for r in poly_data if r["month"] == month), None)
    if row is None:
        return None, None, None, None

    d = day_of_month + (gmt_hh / 24.0)  
    x = d / 32.0

    # 4) Dec & Equation of Time
    Dec_deg = poly_value(row["Dec_a0"], row["Dec_a1"], row["Dec_a2"], row["Dec_a3"], row["Dec_a4"], x)
    E_val   = poly_value(row["E_a0"],  row["E_a1"],  row["E_a2"],  row["E_a3"],  row["E_a4"],  x)
    E_val_h = E_val / 60.0

    # حساب التوقيت الشمسي المحلي
    lon_hours = lon_dd / 15.0
    #local_solar_time = gmt_hh + lon_hours + E_val_h
    #local_solar_time = hours_mod24(local_solar_time)

    gmt_hh = local_hh - utc_corr_hh
    H_g = None
    if E_val is not None and gmt_hh is not None:
        if E_val == False:
            if (gmt_hh + E_val_h) >= 24:
                H_g = gmt_hh + E_val_h - 24
            elif (gmt_hh + E_val_h) < 24:
                H_g = gmt_hh + E_val_h
        else:
            if (gmt_hh + E_val) >= 24:
                H = gmt_hh + E_val - 24
            elif (gmt_hh + E_val) < 24:
                H_g = gmt_hh + E_val

    # Hour Angle
    HA_deg = (H_g+lon_hours)*15   

    # Hour Angle
    #HA_deg = (local_solar_time - 12.32) * 15.0

    lat_r = math.radians(lat_dd)
    dec_r = math.radians(Dec_deg)
    ha_r  = math.radians(HA_deg)

    # Alt
    sin_alt = (math.sin(lat_r)*math.sin(dec_r) +
               math.cos(lat_r)*math.cos(dec_r)*math.cos(ha_r))
    sin_alt = max(-1.0, min(1.0, sin_alt))
    alt_deg = math.degrees(math.asin(sin_alt))

    # Az
    cos_alt = math.cos(math.radians(alt_deg))
    if abs(cos_alt) < 1e-9:
        # قريب من سمت الرأس
        az_deg = 0.0
    else:
        sin_az = (- math.sin(ha_r) * math.cos(dec_r)) / cos_alt
        cos_az = ((math.sin(dec_r) - math.sin(lat_r)*sin_alt) /
                  (math.cos(lat_r)*cos_alt))
        az_rad = math.atan2(sin_az, cos_az)
        az_deg = math.degrees(az_rad)
        az_deg = normalize_0_360(az_deg)

    return az_deg, alt_deg, Dec_deg, E_val,H_g 

# ============= (4) تحضير بيانات الجدول الافتراضي للجايروسكوب ============
default_rows = []
for i in range(1, 31):
    face = "Left" if (i % 2) != 0 else "Right"
    default_rows.append({
        "face": face,
        "local_h": None,
        "local_m": None,
        "local_s": None,
        "dir_ro_deg": None,
        "dir_ro_min": None,
        "dir_ro_sec": None,
        "dir_sun_deg": None,
        "dir_sun_min": None,
        "dir_sun_sec": None,
        "obs_true_heading": None if face == "Left" else "None"
    })

# ============= (5) بناء تطبيق Dash ============
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])
app.title = "Solar Azimuth Observation"

app.layout = dbc.Container([
    # نخزّن معاملات polynomial في Store
    dcc.Store(id="poly-store", data=poly_data),
    
    # إدخال كلمة المرور + زر
    dbc.Row([
        dbc.Col([
            html.Label("Enter Admin Password:"),
            dbc.Input(id="admin-pass", type="password", placeholder="****"),
            dbc.Button("Unlock Polynomials", id="unlock-btn", n_clicks=0, className="mt-2"),
            html.Div(id="unlock-msg", className="text-danger mt-2")
        ], width=3),
    ], className="mb-4"),

    # جدول معاملات Polynomial مخفي حتى إدخال الباسوورد الصحيح
    dbc.Collapse(
        dbc.Card([
            dbc.CardHeader(
                dcc.Input(
                    id="editable-header",
                    value="Monthly Polynomials (Admin Only)",
                    type="text",
                    style={"width": "100%", "border": "none", "fontSize": "1.25rem"}
                )
            ),  # إغلاق CardHeader
            dbc.CardBody([
                dash_table.DataTable(
                    id="poly_data",
                    columns=[
                        {"name": "Month", "id": "month", "type": "numeric", "editable": False},
                        {"name": "RA_a0", "id": "RA_a0", "type": "numeric", "editable": True},
                        {"name": "RA_a1", "id": "RA_a1", "type": "numeric", "editable": True},
                        {"name": "RA_a2", "id": "RA_a2", "type": "numeric", "editable": True},
                        {"name": "RA_a3", "id": "RA_a3", "type": "numeric", "editable": True},
                        {"name": "RA_a4", "id": "RA_a4", "type": "numeric", "editable": True},
                        {"name": "Dec_a0", "id": "Dec_a0", "type": "numeric", "editable": True},
                        {"name": "Dec_a1", "id": "Dec_a1", "type": "numeric", "editable": True},
                        {"name": "Dec_a2", "id": "Dec_a2", "type": "numeric", "editable": True},
                        {"name": "Dec_a3", "id": "Dec_a3", "type": "numeric", "editable": True},
                        {"name": "Dec_a4", "id": "Dec_a4", "type": "numeric", "editable": True},
                        {"name": "E_a0", "id": "E_a0", "type": "numeric", "editable": True},
                        {"name": "E_a1", "id": "E_a1", "type": "numeric", "editable": True},
                        {"name": "E_a2", "id": "E_a2", "type": "numeric", "editable": True},
                        {"name": "E_a3", "id": "E_a3", "type": "numeric", "editable": True},
                        {"name": "E_a4", "id": "E_a4", "type": "numeric", "editable": True},
                        {"name": "SD_a0", "id": "SD_a0", "type": "numeric", "editable": True},
                        {"name": "SD_a1", "id": "SD_a1", "type": "numeric", "editable": True},
                        {"name": "SD_a2", "id": "SD_a2", "type": "numeric", "editable": True},
                        {"name": "SD_a3", "id": "SD_a3", "type": "numeric", "editable": True},
                        {"name": "SD_a4", "id": "SD_a4", "type": "numeric", "editable": True}
                    ],
                    data=poly_data,
                    editable=True,
                ),
            ])
        ], className="mb-4"),
        id="admin-poly-collapse",
        is_open=False
    ),

    # --- بقية أقسام الإدخال (تفاصيل المشروع والجدول) ---
    dbc.Row([
        dbc.Col([
            html.H1("Solar Azimuth Observation", className="text-center"),
            dbc.Input(id="device-name", placeholder="Enter Device Name", type="text", style={"margin-top": "10px"})
        ], width=8),
        dbc.Col([
            html.Img(src="https://via.placeholder.com/150", style={"max-width": "100%"})
        ], width=4, className="text-center")
    ], className="my-4"),

    dbc.Card([
        dbc.CardHeader("Job Details"),
        dbc.CardBody([
            dbc.Row([
                dbc.Col([html.Label("Job Number"), dbc.Input(id="job-number")], width=3),
                dbc.Col([html.Label("Job Description"), dbc.Input(id="job-description")], width=3),
                dbc.Col([html.Label("Client"), dbc.Input(id="client")], width=3),
            ]),
            dbc.Row([
                dbc.Col([html.Label("Surveyor"), dbc.Input(id="surveyor")], width=3),
                dbc.Col([html.Label("Gyro Type and/or ID"), dbc.Input(id="gyro_type_and_or_id")], width=3),
                dbc.Col([html.Label("Vessel"), dbc.Input(id="vessel")], width=3),
                dbc.Col([
                    html.Label("Date"),
                    dcc.DatePickerSingle(id="date", display_format="YYYY-MM-DD")
                ], width=3),
            ]),
            dbc.Row([
                dbc.Col([
                    html.Label("Hemisphere"),
                    dbc.Select(
                        id="hemisphere",
                        options=[{"label": "North", "value": "North"}, {"label": "South", "value": "South"}]
                    )
                ], width=3),
            ])
        ])
    ], className="mb-4"),

    dbc.Card([
        dbc.CardHeader("(Please do not enter any values ​​with a negative sign, just specify whether it is in the Northern or Southern Hemisphere.)"),
        dbc.CardBody([
            html.Label("Latitude (DMS)"),
            dbc.Row([
                dbc.Col(dbc.Input(id="lat-deg", placeholder="Deg", type="number"), width=4),
                dbc.Col(dbc.Input(id="lat-min", placeholder="Min", type="number"), width=4),
                dbc.Col(dbc.Input(id="lat-sec", placeholder="Sec", type="number"), width=4),
            ], className="mb-3"),

            html.Label("Longitude (DMS)"),
            dbc.Row([
                dbc.Col(dbc.Input(id="lon-deg", placeholder="Deg", type="number"), width=4),
                dbc.Col(dbc.Input(id="lon-min", placeholder="Min", type="number"), width=4),
                dbc.Col(dbc.Input(id="lon-sec", placeholder="Sec", type="number"), width=4),
            ], className="mb-3"),

            html.Label("UTC Correction (H:M:S)"),
            dbc.Row([
                dbc.Col(dbc.Input(id="utc-h", type="number", placeholder="±Hours"), width=4),
                
            ], className="mb-4"),
        ])
    ], className="mb-4"),

    dbc.Card([
        dbc.CardHeader(" Solar  ObservationTable (30 rows)"),
        dbc.CardBody([
            dash_table.DataTable(
                id="table_data",
                columns=[
                    {"name": "Face", "id": "face", "editable": False},
                    {"name": "Local Time (H)", "id": "local_h", "type": "numeric", "editable": True},
                    {"name": "Local Time (M)", "id": "local_m", "type": "numeric", "editable": True},
                    {"name": "Local Time (S)", "id": "local_s", "type": "numeric", "editable": True},
                    {"name": "Dir R.O. (Deg)", "id": "dir_ro_deg", "type": "numeric", "editable": True},
                    {"name": "Dir R.O. (Min)", "id": "dir_ro_min", "type": "numeric", "editable": True},
                    {"name": "Dir R.O. (Sec)", "id": "dir_ro_sec", "type": "numeric", "editable": True},
                    {"name": "Dir Sun (Deg)", "id": "dir_sun_deg", "type": "numeric", "editable": True},
                    {"name": "Dir Sun (Min)", "id": "dir_sun_min", "type": "numeric", "editable": True},
                    {"name": "Dir Sun (Sec)", "id": "dir_sun_sec", "type": "numeric", "editable": True},
                    {"name": "Observed True Heading (Left)", "id": "obs_true_heading", "type": "text", "editable": True},
                ],
                data=default_rows,
                style_cell={
                    "textAlign": "center",
                    "whiteSpace": "normal",
                    "minWidth": "80px",
                    "width": "120px",
                    "maxWidth": "180px",
                },
                style_header={"backgroundColor": "rgb(240, 240, 240)", "fontWeight": "bold"},
                editable=True,
            ),
        ])
    ], className="mb-4"),

    dbc.Row([
        dbc.Col([
            dbc.Button("Submit", id="submit-btn", color="primary", n_clicks=0)
        ], width=12, className="text-center"),
    ], className="mb-4"),

    dbc.Row([
        dbc.Col([
            html.H4("Job Details", className="mt-3"),
            html.Div(id="job-details-output", className="mb-4"),

            html.H4("Calculated Results (Pairs)", className="mt-3"),
            html.Div(id="results-table", className="mb-4"),

            html.H4("Statistical Analysis", className="mt-3"),
            html.Div(id="stats-output", style={"fontWeight": "bold", "fontSize": "16px"})
        ])
    ]),
], fluid=True)


# (A) فك قفل جدول البولينومي عند إدخال كلمة مرور
@app.callback(
    Output("admin-poly-collapse", "is_open"),
    Output("unlock-msg", "children"),
    Input("unlock-btn", "n_clicks"),
    State("admin-pass", "value"),
    prevent_initial_call=True
)
def unlock_polynomials(n_clicks, pw):
    if not n_clicks:
        return False, ""
    if pw == "admin123":
        return True, "Polynomials Unlocked (You can now edit)."
    else:
        return False, "Wrong password!"

# (B) تحديث تخزين معاملات Polynomial عند تعديل الجدول
@app.callback(
    Output("poly-store", "data"),
    Input("poly_data", "data"),
    prevent_initial_call=True
)
def update_poly_store(newdata):
    return newdata

# (C) عند الضغط على Submit تحسب الأزيموث وتعرض النتائج
@app.callback(
    [
        Output("results-table", "children"),
        Output("stats-output", "children"),
        Output("job-details-output", "children"),
    ],
    Input("submit-btn", "n_clicks"),
    [
        State("job-number", "value"),
        State("job-description", "value"),
        State("client", "value"),
        State("surveyor", "value"),
        State("gyro_type_and_or_id", "value"),
        State("vessel", "value"),
        State("date", "date"),
        State("hemisphere", "value"),
        State("lat-deg", "value"),
        State("lat-min", "value"),
        State("lat-sec", "value"),
        State("lon-deg", "value"),
        State("lon-min", "value"),
        State("lon-sec", "value"),
        State("utc-h", "value"),
        State("table_data", "data"),
        State("poly-store", "data"),
        State("device-name", "value")
    ]
)

def process_pairs(
    n_clicks,
    job_number, job_desc, client, surveyor,
    gyro_type_and_or_id, vessel, 
    date_val, hemisphere,
    lat_deg, lat_min, lat_sec,
    lon_deg, lon_min, lon_sec,
    utc_h,
    table_data, poly_data, device_name
):
    if not n_clicks:
        return "", "", ""

    # تحقق من صحة التاريخ
    if not date_val:
        return "Please select a valid date.", "", ""

    try:
        dt = datetime.datetime.strptime(date_val, "%Y-%m-%d")
        month = dt.month
        day_of_month = dt.day
    except:
        return "Invalid date format", "", ""

    # خط العرض والطول
    lat_dec = dms_to_decimal(lat_deg, lat_min, lat_sec)
    lon_dec = dms_to_decimal(lon_deg, lon_min, lon_sec)
    if lat_dec is None or lon_dec is None:
        return "Invalid Latitude/Longitude", "", ""

    if hemisphere == "South":
        lat_dec = -abs(lat_dec)
    else:
        lat_dec = abs(lat_dec)

    utc_corr_hh = None
    if utc_h is not None:
        utc_corr_hh = float(utc_h)
    else:
        utc_corr_hh = 0.0


    if not table_data or len(table_data) < 20:
        return "Data table is incomplete.", "", ""

    # قسّم إلى أزواج
    pairs = []
    for i in range(0, 20, 2):
        L = table_data[i]
        R = table_data[i + 1]
        pairs.append((L, R))

    results = []
    co_diffs = []
    local_t_secs = []

    pair_idx = 0
    for (Lrow, Rrow) in pairs:
        pair_idx += 1
        # متوسط الوقت
        LS = time_to_seconds(Lrow.get("local_h"), Lrow.get("local_m"), Lrow.get("local_s"))
        RS = time_to_seconds(Rrow.get("local_h"), Rrow.get("local_m"), Rrow.get("local_s"))
        if (LS is not None) and (RS is not None):
            avgS = 0.5 * (LS + RS)
        else:
            avgS = LS if LS is not None else RS

        avgTime_str = "N/A"
        local_hh_decimal = None
        if avgS is not None:
            conv = seconds_to_hms(avgS)
            if conv:
                hh, mm, ss = conv
                local_hh_decimal = hh + mm/60.0 + ss/3600.0
                avgTime_str = f"{hh:02d}:{mm:02d}:{ss:02d}"

            local_t_secs.append(avgS)

        # حساب الفرق الأفقي
        def Hangle(rr):
            roD = dms_to_decimal(rr.get("dir_ro_deg"), rr.get("dir_ro_min"), rr.get("dir_ro_sec"))
            suD = dms_to_decimal(rr.get("dir_sun_deg"), rr.get("dir_sun_min"), rr.get("dir_sun_sec"))
            return None if (roD is None or suD is None) else (suD)

        haL = Hangle(Lrow)
        haR = Hangle(Rrow)

        avg_deg = None

        if haL is not None and haR is not None:
            if haL > haR:
                avg_deg = (haL + (haR + 180)) / 2.0
            else:
                avg_deg = (haL + (haR - 180)) / 2.0

        if avg_deg is not None:
            arcseconds_float = avg_deg * 3600.0
            arcseconds_rounded = round(arcseconds_float)
            avg_ha = arcseconds_rounded / 3600.0
            avg_ha_str = decimal_to_dms(avg_ha)
        else:
            avg_ha_str = "N/A"

        sunAz_str = roAz_str = calcTH_str = obs_str = dco_str = "N/A"

        if local_hh_decimal is not None:
            try:
                az_deg, alt_deg, dec_deg, e_val, H = get_azimuth(
                    lat_dec, lon_dec, month, day_of_month,
                    local_hh_decimal, utc_corr_hh, poly_data
                )
                if az_deg is not None:
                    sunAz_str = decimal_to_dms(az_deg)
                    if avg_ha is not None:
                        ro_az = az_deg - avg_ha
                        ro_az = normalize_0_360(ro_az)
                        roAz_str = decimal_to_dms(ro_az)
                        calcTH_str = f"{ro_az-180:.4f}"
            except Exception as e:
                print("Azimuth Error:", e)

        # هنا نطبق المعادلة الجديدة لحساب (C - O)
        if Lrow["face"] == "Left":
            val = Lrow.get("obs_true_heading")
            if val and val not in ["None", ""]:
                try:
                    obs_val = float(val)
                    obs_str = f"{obs_val:.4f}"

                    if calcTH_str != "N/A":
                        cth = float(calcTH_str)
                        dco = cth - obs_val
                        
                        if abs(dco) >= 200:
                            dco += 360  # إذا كان الفرق أكبر من أو يساوي 200 درجة، أضف 360
                        
                        dco_str = f"{dco:.4f}"
                        co_diffs.append(dco)
                except:
                    pass

        results.append({
            "Pair": pair_idx,
            "AvgLocalTime": avgTime_str,
            "AvgHorizAngle(DMS)": avg_ha_str,
            "SunAz(DMS)": sunAz_str,
            "R.O.Az": roAz_str,
            "calculated true heading": calcTH_str,
            "observation true heading": obs_str,
            "C-O": dco_str
        })

    cols = ["Pair", "AvgLocalTime", "AvgHorizAngle(DMS)", "SunAz(DMS)", "R.O.Az", "calculated true heading", "observation true heading", "C-O"]
    if results:
        thead = html.Thead(html.Tr([html.Th(c) for c in cols]))
        tbody = html.Tbody([
            html.Tr([html.Td(str(rr[c])) for c in cols]) for rr in results
        ])
        table_res = html.Table([thead, tbody], className="table table-bordered table-striped")
    else:
        table_res = html.Div("No pair results.")

    # إحصائيات
    avg_t_str = "N/A"
    if local_t_secs:
        mean_s = sum(local_t_secs) / len(local_t_secs)
        conv = seconds_to_hms(mean_s)
        if conv:
            HH, MM, SS = conv
            avg_t_str = f"{HH:02d}:{MM:02d}:{SS:02d}"

    st_list = [f"Overall Avg Local Time = {avg_t_str}"]
    if co_diffs:
        mean_co = np.mean(co_diffs)
        std_co = np.std(co_diffs)
        st_list.append(f"Mean(C - O) = {mean_co:.4f}°")
        st_list.append(f"StdDev(C - O) = {std_co:.4f}°")

    stats_div = html.Ul([html.Li(s) for s in st_list])

    job_info = {
        "Job Number": job_number or "",
        "Job Description": job_desc or "",
        "Client": client or "",
        "Surveyor": surveyor or "",
        "Gyro Type and/or ID": gyro_type_and_or_id or "",
        "Vessel": vessel or "",
        "Date": date_val,
        "Hemisphere": hemisphere or "",
        "Latitude (DMS)": f"{lat_deg or ''}° {lat_min or ''}' {lat_sec or ''}\"",
        "Longitude (DMS)": f"{lon_deg or ''}° {lon_min or ''}' {lon_sec or ''}\"",
        "Device Name": device_name or "",
    }

    job_table = html.Table([
        html.Tbody([
            html.Tr([html.Th(k), html.Td(str(v))]) for k, v in job_info.items()
        ])
    ], className="table table-bordered table-striped")

    return table_res, stats_div, job_table


if __name__ == '__main__':
    app.run_server(debug=True) 

    





 
```
